```python
ratings = open('data/ratings.csv').readlines()


ratings = ratings[1:]       # removing the header
print(ratings[:4])
```

```python
user_to_books = {}          # {user: {book0: rating, book1: rating, ...}}

# turns out this isn't really useful, but I'll leave it here 
book_to_users = {}          # {book: {user0: rating, user1: rating, ...}}
```

```python
for line in ratings:
    user_id, book_id, rating = line.split(',')
    rating = int(rating)
    
    if user_id in user_to_books.keys():
        # this assumes there aren't multiple ratings for a user-book pair, which there shouldn't 
        # might be better to check first though
        user_to_books[user_id][book_id] = rating
    else: 
        user_to_books[user_id] = {book_id: rating}
        
    if book_id in book_to_users.keys():
        book_to_users[book_id][user_id] = rating
    else:
        book_to_users[book_id] = {user_id: rating}
```

#### To build the dataset in a way that it is efficient to be used later, we will explicitly save all the pairs so we will avoid picking from the data during training with 'random' and stuff that would slow it down 

```python
# calculating dataset lengths 

def factorial(x):
    if x == 1:
        return x
    else:
        return x * factorial(x-1)
    
    
def pick(n, k):
    return factorial(n) / (factorial(n-k) * factorial(k))


pos_book_user_length = 0
pos_book_book_length = 0
for user in range(user_num):
    n_books = len(user_to_books[str(user + 1)].keys())  
    pos_book_user_length += n_books      # +1 because the first user_id is 1
    # just summing these as for each user and each book we have a pair 
    pos_book_book_length += pick(n_books, 2)
    
print('positive BOOK_USER pairs are ', pos_book_user_length)
print('positive BOOK_BOOK pairs are', pos_book_book_length)
```

```python
import csv
```

```python
with open('user_book.csv', 'w') as writeFile:
    writer = csv.writer(writeFile)
    pairs = 0
    for user in range(user_num):
        user_id = str(user+1)
        this_user_to_books = user_to_books[user_id]
        for book in this_user_to_books.keys():
            if this_user_to_books[book] > 3:            # considering only ratings above 3
                writer.writerow([user_id, book])
                pairs += 1
                
        if user % 10000 == 0:
            print(user)
                
writeFile.close()
print('There are', pairs, 'user_book pairs')
```

```python
with open('book_book.csv', 'w') as writeFile:
    writer = csv.writer(writeFile)
    pairs = 0
    # if n users have rated the same two books twice, that pair will be added n times
    # should this be the case? 
    for user in range(user_num):
        user_id = str(user+1)
        this_user_to_books = user_to_books[user_id]
        for i in range(len(this_user_to_books.keys())):
            for book1 in list(this_user_to_books.keys())[i+1:]:
                book0 = list(this_user_to_books.keys())[i]
                
                # still considering only ratings above 3
                if this_user_to_books[book0] > 3 and this_user_to_books[book1] > 3:
                    writer.writerow([list(this_user_to_books.keys())[i], book1])
                    pairs += 1
        
        if user % 1000 == 0:
            print(user)
            
writeFile.close()
print('There are', pairs, 'book_book pairs')
```
